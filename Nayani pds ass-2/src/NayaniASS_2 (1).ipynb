
Imputing missing values with the mode is justified when dealing with categorical or ordinal data like the number of seats in a dataset. It ensures that no data is lost and helps maintain the integrity of the dataset by replacing missing values with the most frequently occurring value.

```python
#seats mode
```

```python
seats_mode = df['Seats'].mode()[0]
seats_mode
```

```python
df['Seats'].fillna(seats_mode, inplace=True)
df
```

```python
missing_values = df.isnull().sum()
missing_values
```


Imputing missing values in the "Mileage" column with the median is justified because mileage data typically follows a skewed distribution. Using the median as a measure of central tendency is robust to outliers and provides a representative value. Additionally, it preserves the integrity of the data and ensures consistency in analysis.
